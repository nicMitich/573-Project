```python
def extract_text_from_txt(txt_path):        # Function to read the content of a text file and return it as a string
    with open(txt_path, 'r', encoding='utf-8') as file:     # Open the specified text file in read mode with UTF-8 encoding
        return file.read()

sample_text = extract_text_from_txt('My Resume-intext.txt')     # Extract the text content from the specified text file and store it in the variable sample_text
print(sample_text[:1000])
```

```python
import re       # For regular expressions to clean the text

def clean_text(text):
    text = re.sub(r'\n+', '\n', text)  # Replace multiple newlines with a single newline   
    # \n = newline, + = 1/more so \n+ = 1/more newlines
    text = re.sub(r' +', ' ', text)  # Replace multiple spaces with a single space
        # ' ' = space, + = 1/more so ' '+ = 1/more spaces
    return text.strip()

cleaned_text = clean_text(sample_text)      # Clean the extracted text to make it easier to parse and use
print(cleaned_text)     # Print the cleaned text to verify the cleaning process
```

```python
def extract_email(text):        # Extract email addresses using a regular expression
    match = re.search(r'\S+@\S+', text)     # \S = non-whitespace, + = 1/more so \S+ = 1/more non-whitespace, @ = @ symbol
    return match.group(0) if match else None        # Return the matched email address if found, otherwise return None

def extract_phone_number(text):         # Extract phone numbers using a regular expression
    match = re.search(r'\+?\d[\d -]{8,}\d', text)       # \+? = optional + at the start, \d = digit, [\d -]{8,} = 8 or more digits/spaces/hyphens, \d = digit at the end
    return match.group(0) if match else None    # Return the matched phone number if found, otherwise return None

email = extract_email(cleaned_text)     # Extract the email address from the cleaned text
phone_number = extract_phone_number(cleaned_text)       # Extract the phone number from the cleaned text
print(f"Extracted Email: {email}")
print(f"Extracted Phone Number: {phone_number}")        # Print the extracted email and phone number to verify the extraction process
```

```python
import spacy        # For natural language processing and entity recognition to extract the name from the resume
# spaCy NER is the Named Entity Recognition feature within the spaCy open-source 
# library for Natural Language Processing (NLP) in Python

nlp = spacy.load('en_core_web_sm')

def extract_name(text):     # Use spaCy's named entity recognition to extract the person's name from the text
    doc = nlp(text)
    for ent in doc.ents:
        if ent.label_ == 'PERSON':  
            return ent.text
    return None

name = extract_name(cleaned_text)
print(f"Extracted Name: {name}")        # Print the extracted name to verify the extraction process
```

```python
EDUCATION_KEYWORDS = ['Associate', 'Associates', 'Bachelor', 'Master', 'Bachelors', 'Masters', 'B.Sc', 'M.Sc', 'PhD', 'B.E', 'M.E.']

def extract_education(text):
    lines = text.split('\n')     # Split the text into lines to check for education-related keywords in each line
    education_info = []
    for line in lines:
        if any(keyword in line for keyword in EDUCATION_KEYWORDS):     # Check if any of the education-related keywords are present in the line
            education_info.append(line.strip())     # If a line contains any of the education keywords, add it to the education_info list after stripping leading/trailing whitespace
    return education_info

education_details = extract_education(cleaned_text)  # Extract education details from the cleaned text and store them in a list
print("Extracted Education Details:")
for edu in education_details:
    print("-", edu)    # Print each extracted education detail to verify the extraction process
```

```python
def extract_experience(text):
    experience_keywords = ['experience', 'work', 'employment', 'internship']  # Keywords to identify lines related to work experience in the resume
    experience_lines = []
    lines = text.split('\n')
    for line in lines:
        if any(keyword in line.lower() for keyword in experience_keywords):  # Check if any of the experience-related keywords are present in the line (case-insensitive)
            experience_lines.append(line.strip())
    return experience_lines

experience_details = extract_experience(cleaned_text)  # Extract experience details from the cleaned text and store them in a list
print("Extracted Experience Details:")
for exp in experience_details:
    print("-", exp)   # Print each extracted experience detail to verify the extraction process
```

```python
parsed_resume = {     # Create a dictionary to store all the extracted information from the resume in a structured format
    'name': extract_name(cleaned_text),
    'email': extract_email(cleaned_text),
    'phone_number': extract_phone_number(cleaned_text),
    'skills': extract_skills(cleaned_text),
    'education': extract_education(cleaned_text),
    'experience': extract_experience(cleaned_text)
}

print("Parsed Resume:")    # Print the structured dictionary containing all the extracted information from the resume
for key, value in parsed_resume.items():
    print(f"{key.capitalize()}: {value}")
```

```python
import pandas as pd     # For handling data in a structured format and exporting to CSV
import os

def process_folder(folder_path):        # Process all text files in the specified folder, extract relevant information from each resume, and compile it into a DataFrame
    resumes_data = []
    for filename in os.listdir(folder_path):    # Iterate through all files in the specified folder to find and process text files containing resumes
        if filename.endswith('.txt'):
            txt_path = os.path.join(folder_path, filename)
            text = extract_text_from_txt(txt_path)
            cleaned_text = clean_text(text)
            parsed_resume = {
                'name': extract_name(cleaned_text),
                'email': extract_email(cleaned_text),
                'phone_number': extract_phone_number(cleaned_text),
                'skills': extract_skills(cleaned_text),
                'education': extract_education(cleaned_text),
                'experience': extract_experience(cleaned_text)
            }
            resumes_data.append(parsed_resume)  # Append the structured dictionary for each resume to the resumes_data list#
    return pd.DataFrame(resumes_data)

df = process_folder('resumes')  # Process the folder containing the resume text files and create a DataFrame with the extracted information from all resumes
df.to_csv('parsed_resumes.csv', index=False)
```
